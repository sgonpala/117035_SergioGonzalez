```python
"""Lee una lista de números ordenados de forma ascendente, verifica que esté 
ordenada, lee otro número y lo ubica en la posicicón adecuada"""


class A(Exception):                               # Definición de la excepción
    """La lista no se encuentra ordenada."""
    pass


while True:
    try:
        lista1 = eval(input("Ingrese una lista de números ordenada: "))  
                                                     # Solicitud de la lista
        for i in range(1, len(lista1) - 1):
            if lista1[i - 1] <= lista1[i] <= lista1[i + 1]:# Análisis de orden
                pass
            else:
                raise A
        else:
            break

    except A:
        print("LA LISTA NO ESTÁ ORDENADA.")

while True:
    try:                              # Solicitud del elemento para ordenar
        ele = float(input("Ingrese un número para insertar en la lista: "))
        print()
        break
    except ValueError:
        print("INGRESE UN NÚMERO ADECUADO.")


for i in range(len(lista1)):       # Insertar el elemento en el orden adecuado
    if lista1[i] > ele:
        lista1.insert(i, ele)
        break
else:
    lista1.append(ele)

print("La lista con el elemento adicional en la posición correcta es:", lista1)



```

```python
'''Leer una lista y determinar cuál es el segundo elemento más grande de dicha
 lista. No utilice ningún tipo de algoritmo de ordenamiento para realizar este
 ejercicio.'''
 
LIST = []      # Creación de la lista
size = int(input('Ingrese el tamaño de la lista: ')) # Solicitud del tamaño de
                                                     # la lista

while size < 0:      # Si el valor es negativo, pedir valor positivo
    print('Por favor ingrese un valor positivo.') 
    size = int(input('Ingrese el tamaño de la lista: ')) # Nueva solicitud de
                                                         # lista
else:                                                                              
    for elemento in range(size): # Creación de un ciclo 
        LIST.append(input(f'Ingrese el elemento {elemento + 1}: '))
        
    maxi = max(LIST) # Máximo elemento de la lista y se almacena
    LIST.remove(maxi) # Se elimina de la lista el elemento mayor
    val = max(LIST) # Nuevo valor máximo de la lista
    print(f'El segundo elemento más grande de la lista es {val}')
```
